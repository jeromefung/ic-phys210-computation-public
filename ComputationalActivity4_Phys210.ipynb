In today's activity, we will examine two complementary ways to visualize solutions to first-order ODEs. 

One approach is to plot the analytic solutions to a given ODE for different values of the arbitrary constant. This technique is readily generalized to second-order equations. It only works, however, if you have already found an analytic solution. Nonetheless, much of the *physics* arguably lies not so much in *finding* the solutions (which can often be outsourced to handbooks, software, or mathematicians) but in *interpreting* the solutions, and this approach can be quite helpful in making sense of the solutions. You've already learned how to plot functions in a previous activity, so this approach requires no new computational tools.

The second approach is to plot the *slope field* for a given differential equation. (See Sec. 1.4 of Felder & Felder for more details.) Looking at a slope field can give you insight into the behavior of solutions to a first-order ODE *without* knowing or being able to find the solutions themselves. In fact, given a slope field and an initial condition, you can sketch what the particular solution must be. 

We will use matplotlib to plot slope fields. This is certainly not the only or easiest tool you could use, but it will build on what you've already done. In addition, the techniques you learn today for plotting vector fields could be readily adapted to visualizing other vector fields that arise in physics, such as electric and magnetic fields. ## Part 2: Slope fields

In this section we will consider the slope fields for the two differential equations we examined in Part 1. As discussed in Felder and Felder, a slope field is a *two-dimensional* plot: we generate a grid of $x$ and $y$ values and plot, at each point on the grid, a short line segment or arrow whose slope is $dy/dx$.

Let's consider the differential equation for the charging capacitor again:

$$\frac{dQ}{dt} = 2 - \frac{2}{3}Q.$$

We'll plot $t$ on the horizontal axis and $Q$ on the vertical axis, as in Figure 1.2 of Felder & Felder. The first thing we have to do is generate a grid of points. Here's the easiest way to do it:

```python
ts, Qs = np.meshgrid(np.arange(0, 6, 0.5), np.arange(0, 10, 1))
```

The line above generated two arrays, `ts` and `Qs`, that represent grids of the same shape. As you can see below, in the `ts` grid, all the values in a given column are the same. Each separate column represents a value of $t$ from 0 to 5.5 s in steps of 0.5 s. Similarly, in the `Qs` grid, all the values in a given row are the same, and each row represents a value of $Q$ from 0 to 9 C in steps of 1 C.

```python
print(ts)
```

```python
print(Qs)
```

Now that we've clarified what the `np.meshgrid` command does, let's modify the definitions of `ts` and `Qs` in the cell below so that *both* run from 0 to 6 (excluding endpoints) in steps of 0.5 (in the appropriate units).

```python
ts, Qs = np.meshgrid(np.arange(0, 6, 0.5), np.arange(0, 6, 0.5))
```

We can plot a slope field using the `matplotlib.pyplot.quiver` command. (Look up the documentation.) To use this command, we need to specify a minimum of 4 things: the $x$ and $y$ coordinates of the grid point, and then at each grid point, the run $\Delta x$ and the rise $\Delta y$ of the slope at each point. `quiver` is primarily intended for plotting vector fields where one might want to specify the $x$ and $y$ components of the vector separately. 

Here's the strategy we'll use:

* Calculate `dx`. We'll set this to 1, and this won't change.
* Calculate `dy`. Because we've set `dx` to one, this is the derivative $dy/dx$.
* We want to plot normalized arrows or lines, so we'll calculate the magnitude $\sqrt{dx^2 + dy^2}$ and then normalize both $dx$ and $dy$ by this value.

Here's how this will work:

```python
dx = np.ones_like(ts) 
# np.ones_like is a convenience function that outputs an array of 1s in the same shape as the input array.
```

To show you how this works, I'll draw the slope field for the differential equation

$$\frac{dQ}{dx} = Q -1,$$

and I'll ask you to modify for the differential equation we actually care about afterwards.

```python
dy = Qs - 1
magnitude = np.sqrt(dx**2 + dy**2)
```

Now we can make the plot:

```python
plt.quiver(ts, Qs, dx/magnitude, dy/magnitude, pivot = 'mid', angles = 'xy')
# First argument is grid of x values, second argument is grid of y values, third is array of runs Delta x, and fourth
# is array of rises Delta y
```

There were two keyword arguments. pivot = `mid` drew the middle of each arrow over each grid point (as opposed to the tail, which is the default). angles = `xy` is recommended for plotting slope fields because it keeps the directions of the arrows consistent if the scaling for the $x$ and $y$ axes is different.

You can also, if you prefer, plot line segments (as in Felder and Felder) rather than arrows by adding two more keyword arguments:

```python
plt.quiver(ts, Qs, dx/magnitude, dy/magnitude, pivot = 'mid', angles = 'xy', headwidth = 1, headlength = 0)
```

Look up the documentation for `matplotlib.pyplot.quiver` if you want more details about how those last two keyword arguments work. It's totally up to you whether you want to plot lines or arrows.
