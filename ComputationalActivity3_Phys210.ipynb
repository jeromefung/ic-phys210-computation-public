## Plotting with matplotlib

Python, by itself, does not have any built-in plotting abilities. Neither does the `numpy` library for working with arrays, which you began using last time. However, there is a now-standard library for making 2D plots (as well as rudimentary 3D plots) called `matplotlib`, which is designed to work with numpy arrays.

There are two user interfaces for `matplotlib`: the simpler-to-use `matplotlib.pyplot` interface and a more powerful, but harder to use, object-oriented interface. In this course we'll only use the simpler `matplotlib.pyplot` interface. Here is a link to the official tutorial for pyplot: [https://matplotlib.org/stable/tutorials/pyplot.html#sphx-glr-tutorials-pyplot-py](https://matplotlib.org/stable/tutorials/pyplot.html#sphx-glr-tutorials-pyplot-py)

Let's get started by revisting the example from the end of the last tutorial. First, we'll need to import the packages we want to use:

```python
import numpy as np
import matplotlib.pyplot as plt
```

In the first line, `import numpy` would be sufficient, but the `as np` allows us to type `np` as a shorthand for `numpy`, making syntax like `np.sin()` valid. Similarly, the `as plt` in the second line saves us from having to type out `matplotlib.pyplot` every time we use a function from that library.

The next thing we'll do is call a *magic function* that only works within Jupyter notebooks. (It's not legal Python syntax otherwise.) However, if you're working within a Jupyter notebook, the following magic function causes the results of plotting commands to be displayed within the notebook, which is often convenient.

```python
%matplotlib inline
```

Our first plot was of $\sin \theta$. To make it, we first had to create an array of angles:

```python
thetas = np.linspace(0, 2*np.pi, 101)
```

In general, plotting only works if you have two arrays (one for the horizontal coordinate, and the other for the vertical coordinate) *of the same size*. (In our experience, errors that occur when you try to generate a plot often arise because of mismatched array sizes.) To avoid trouble, we can use the array `thetas` we just created to calculate an array of values corresponding to the sine of those angles:

```python
sines = np.sin(thetas)
```

You can use the `print()` function to inspect arrays if you want:

```python
print(thetas)
```

```python
print(sines)
```

And now, we can use pyplot `plot` command to make a basic plot. Two arguments are needed:

* First argument: the array containing coordinates to be plotted on the horizontal axis
* Second argument: the array containing coordinates to be plotted on the vertical axis

```python
plt.plot(thetas, sines) 
```

You can, if you want, calculate the sine and plot in one line:

```python
plt.plot(thetas, np.sin(thetas))
```

If no other arguments are given to `plt.plot()`, the points are joined by continuous line segments. This may not always be desirable, and the behavior can be changed by an optional third argument:

```python
plt.plot(thetas, np.sin(thetas), 'o')
```

Other shapes for the plot markers are possible, as you can see in the following example. (This next cell also shows how you can plot more than one thing on the same axes).

```python
plt.plot(thetas, np.sin(thetas), '.') # smaller dots
plt.plot(thetas, np.cos(thetas), '^') # triangles pointing up
plt.plot(thetas, np.sin(thetas/2), 's') # squares
plt.plot(thetas, np.cos(2*thetas), 'v-') # downward-pointing triangles, with lines between the markers
```

Here is a full list of possible markers: [https://matplotlib.org/stable/api/markers_api.html](https://matplotlib.org/stable/api/markers_api.html)

You might also be wondering about the colors. By default, pyplot cycles through a color palette that displays well on most devices. But you can also specify your own colors:

```python
plt.plot(thetas, np.sin(thetas), 'm.') # magenta dots
plt.plot(thetas, np.cos(thetas), 'k>') # black right-pointing triangles
```

If you want, here is more information about specifying colors of your choosing: 
[https://matplotlib.org/stable/users/explain/colors/colors.html#colors-def](https://matplotlib.org/stable/users/explain/colors/colors.html#colors-def)

We'll now show off some other potentially useful features, particularly for making plots in the context of physics or math:

```python
plt.figure(figsize=(8,6)) # make plots larger (nominal size in inches)
# must be called before plotting
# actual on-screen size may vary depending on your computer browser
plt.plot(thetas, np.sin(thetas)) 
plt.xlim((1, 2)) # change x-axis limits
plt.ylim((-2, 12)) # change y-axis limits
plt.xlabel('String with some actual label') # label the x-axis
plt.ylabel('Some other appropriate label', size=12) # label the y-axis
# keyword argument above increases the font size to 12 pt
plt.title('Give your plots better titles than this') # title
```

## Saving plots

Not infrequently, if you're doing research or lab work, you'll want to get your plots *out* of a Jupyter notebook and into a format you can print out and glue into a lab notebook, or include in a Word or LaTeX document. (Screenshots of a Jupyter notebook itself tend to be low-resolution and are *not* recommended, especially if the plot is going to go in a paper, presentation, or poster.) Fortunately, `pyplot` has a handy `savefig` command that you can call after generating a plot. We'll demonstrate in the example below:

```python
plt.plot(thetas, np.sin(thetas))
plt.xlabel(r'$\theta$ [rad]') # using LaTeX syntax
plt.ylabel(r'$\sin\theta$')
plt.savefig('sine_plot.pdf') # the extension specifies the file type
```

After running the cell, you should find a file called `sine_plot.pdf` in the same directory where this notebook lives. PDF is generally the easiest format to use for papers or talks, but on most platforms other formats including PNG or TIFF are also supported.

Note that if you re-run the cell (e.g., after changing it), but don't update the file name, `sine_plot.pdf` will be overwritten.
